```python
def single_or_multi(type):
    if "/" in type:
        return "Multi"
    return "Single"

#위의 데이터에서 "/" 형태를 확인 후 2가지 이상이 있는 경우 Multi로 변환하고 1가지만 있는 경우에는 Single로 변환하는 명령어 이다.
```

```python
pokemons.apply(single_or_multi)
```

```python
pokemons.apply(single_or_multi).value_counts()
#apply를 통해서 변환시킨 값의 개수를 구하시오
```

```python
import seaborn as sns
titanic=sns.load_dataset("titanic")
#tiatnic data를 가지고 오는 방법
```

```python
titanic['adult/child'] = titanic.apply(lambda r: "adult" if r.age >= 20 else "child", axis=1)
# 20살을 기준으로 adult와 child를 구분해서 입력하는 방법
#lambda 매게변수: 표현식(다음과 같은 순서로 lambda를 사용하면 된다.)
titanic.tail()
```

```python
titanic['category'] = titanic.apply(lambda r: "child" if r.age < 20 else r["sex"], axis=1)
# lambda함수는 매게변수를 이용한 리턴 값이다.
# 람다 함수는 나나의 매개 변수를 사용할 수 있지만 여러 개의 매게 변수를 사용하는 것도 가능하다.
# 람다 함수는 리스트를 정렬하여 새로운 리스트로 만드는 내장 함수입니다.

# 대괄호[ ] 는 list에서 사용 -> 수정이 가능한 데이터
# 소괄호( ) 는 튜플에서 사용 -> 수정이 불가능한 데이터
# 중괄호{ } 는 딕셔너리 사용 -> kwy와 value의 쌍의 형태로 이루어진 값으로 수정 가능한 값이다.

# 위의 함수에서 ['sex']를 사용한 이유는 1.sex는 column의 한종류이다/ 2. sex의 하위 범주는 list이다 따라서 수정이 가능한 데이터 이다.
 # 즉 소괄호를 사용하는 튜플과 같이 수정이 불가능하지도 않고, 중괄호를 사용하는 딕셔너리의 자료 형태가 아니기 때문이다.
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan,4]],
                  columns=list("ABCD"))

df
```

```python
df.fillna(0)
# NaN으로 표현된 모든 값들이 0으로 바뀐 것을 확인 할 수 있다.
```

```python
values = {"A": 0, "B" : 1, "C" : 2, "D" : 3}
df.fillna(value=values)
# NaN에 채우고 싶은 값을(결측치를 채우는 값) 정해서 만들 수 있다.
```

```python
values = {"A": 0, "B" : 1, "C" : 2, "D" : 3}
df.fillna(value=values, limit=1)
# column마다 결측치를 변경하고 싶을 때 1번만 사용이 가능하다.
```

```python
df2 = pd.DataFrame(np.zeros((3, 4)), columns=list("ABCD"))
df.fillna(df2)
# 3행의 D열의 NaN 값을 df2에서 가져다가 채운다는 의미의 코딩 함수이다.
```
